```python
def get_note_event(start, note, instrument, length, velocity):
  """This function gets the note event for a particular instrument in the 
  format that the model generates the MIDI text"""
  out_instrument = "%"
  match instrument:
    case 2:
      out_instrument = "%"
    case 3:
      out_instrument = "^"
    case 4:
      out_instrument = "&"
    case 5:
      out_instrument = "*"
    case 6:
      out_instrument = "("
    case 7:
      out_instrument = "("
    case _:
      out_instrument = "%"


  dynamic = "$"
  if(velocity > 108):
    dynamic = "$"
  elif velocity > 74:
    dynamic = "#"
  elif velocity > 40:
    dynamic = "@"
  else:
    dynamic = "!"

  return [start, dynamic, length, out_instrument, note]
```

```python
def parse_split_csv_item(item):
  """This functions splits the MIDI stored in the csv based on 
  whether it is a note, a chord, the velocity, instrument, or duration"""
  try:
    item[2] = item[2].replace(' ', '')
    item[0] = int(item[0])
    item[1] = int(item[1])
    item[3] = int(item[3])
    item[4] = int(item[4])
    item[5] = int(item[5])
  except:
    pass;
  return item
```

```python
def generateMusic(prompt, temperature):
  """This function generates music based on the temperature and prompt provided"""
  generated = torch.tensor(tokenizer.encode(prompt)).unsqueeze(0)
  generated = generated.to(device)

  sample_outputs = model.generate(
                                  generated,
                                  #bos_token_id=random.randint(1,30000),
                                  do_sample=True,
                                  top_k=50,
                                  max_length = 1000,
                                  top_p=1.0,
                                  temperature=temperature,
                                  num_return_sequences=1,
                                  pad_token_id=tokenizer.pad_token_id,
                                  max_new_tokens=200
                                  )

  #return tokenizer.decode(sample_outputs, skip_special_tokens=True)
  for i, sample_output in enumerate(sample_outputs):
    return tokenizer.decode(sample_output, skip_special_tokens=True)
    print("{}: {}\n\n".format(i, tokenizer.decode(sample_output, skip_special_tokens=True)))
  return ""
```

```python
#@title Convert to Midi
reset_midi()

def generate_on_off_events(noteData):
  """This function generates the track or the instrument when there is a change in the note data """
  track_num = 2
  match noteData[3]:
    case "%":
      track_num = 2
    case "^":
      track_num = 3
    case "%":
      track_num = 4
    case "^":
      track_num = 5
    case "&":
      track_num = 6
    case "*":
      track_num = 7
    case "(":
      track_num = 8
    case ")":
      track_num = 9
  #TODO:

  dynamic = 127
  match noteData[1]:
    case "$":
      dynamic = 127
    case "#":
      dynamic = 80
    case "@":
      dynamic = 55
    case "!":
      dynamic = 32

  note_on = str(track_num) + ", " + str(noteData[0]) + ", Note_on_c, 0, " + str(noteData[4]) + ", " + str(dynamic)
  note_off = str(track_num) + ", " + str(noteData[2] + noteData[0]) + ", Note_off_c, 0, " + str(noteData[4]) + ", " + "0"
  return [[noteData[0], note_on, track_num], [noteData[0] + noteData[2], note_off, track_num]]


def test_if_num(num):
  if(num == "-"):
    return True
  try:
    tmp = int(num)
  except:
    return False

  return True

```
